# EDA Visualizations
This notebook creates comprehensive visualizations for exploratory data analysis of the Telco Customer Churn dataset.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Create Results/figures directory if it doesn't exist
import os
figures_dir = r'D:\EDU\DAS 601 ML\Final Project\Results\figures'
os.makedirs(figures_dir, exist_ok=True)
print(f"Figures will be saved to: {figures_dir}")
```

```python
# Load the training dataset
train_path = r'D:\EDU\DAS 601 ML\Final Project\Data\output\train.csv'
df = pd.read_csv(train_path)
print(f"Training dataset loaded with shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

```python
# 1. Bar chart: Gender distribution
plt.figure(figsize=(8, 6))
gender_counts = df['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(gender_counts.values):
    plt.text(i, v + 20, str(v), ha='center', fontweight='bold')

plt.tight_layout()
save_path = r'D:\EDU\DAS 601 ML\Final Project\Results\figures\01_gender_distribution.png'
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")
plt.show()
```

```python
# 2. Grouped bar: Contract vs Churn
plt.figure(figsize=(10, 6))
contract_churn = pd.crosstab(df['Contract'], df['Churn'])
contract_churn.plot(kind='bar', color=['lightgreen', 'salmon'])
plt.title('Contract Type vs Churn', fontsize=16, fontweight='bold')
plt.xlabel('Contract Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
save_path = r'D:\EDU\DAS 601 ML\Final Project\Results\figures\02_contract_vs_churn.png'
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")
plt.show()
```

```python
# 3. Stacked and 100% stacked bar: InternetService vs Churn
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Stacked bar
internet_churn = pd.crosstab(df['InternetService'], df['Churn'])
internet_churn.plot(kind='bar', stacked=True, ax=ax1, color=['lightgreen', 'salmon'])
ax1.set_title('Internet Service vs Churn (Stacked)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Internet Service', fontsize=12)
ax1.set_ylabel('Count', fontsize=12)
ax1.legend(title='Churn', labels=['No', 'Yes'])
ax1.tick_params(axis='x', rotation=45)

# 100% stacked bar
internet_churn_pct = internet_churn.div(internet_churn.sum(axis=1), axis=0) * 100
internet_churn_pct.plot(kind='bar', stacked=True, ax=ax2, color=['lightgreen', 'salmon'])
ax2.set_title('Internet Service vs Churn (100% Stacked)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Internet Service', fontsize=12)
ax2.set_ylabel('Percentage', fontsize=12)
ax2.legend(title='Churn', labels=['No', 'Yes'])
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
save_path = os.path.join(figures_dir, '03_internet_service_vs_churn.png')
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")
plt.show()
```

```python
# 4. Horizontal bar: PaymentMethod counts
plt.figure(figsize=(10, 6))
payment_counts = df['PaymentMethod'].value_counts()
plt.barh(payment_counts.index, payment_counts.values, color='lightblue')
plt.title('Payment Method Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Count', fontsize=12)
plt.ylabel('Payment Method', fontsize=12)
plt.grid(axis='x', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(payment_counts.values):
    plt.text(v + 20, i, str(v), va='center', fontweight='bold')

plt.tight_layout()
save_path = os.path.join(figures_dir, '04_payment_method_distribution.png')
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")
plt.show()
```

```python
# 5. Comprehensive Numeric Variables Analysis: Histograms & Density Plots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Define numeric columns
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

for i, col in enumerate(numeric_cols):
    # Histogram
    axes[0, i].hist(df[col], bins=30, alpha=0.7, color='skyblue', edgecolor='black', density=True)
    axes[0, i].set_title(f'{col} Distribution (Histogram)', fontsize=14, fontweight='bold')
    axes[0, i].set_xlabel(col, fontsize=12)
    axes[0, i].set_ylabel('Density', fontsize=12)
    axes[0, i].grid(alpha=0.3)
    
    # Add statistics text
    mean_val = df[col].mean()
    median_val = df[col].median()
    std_val = df[col].std()
    axes[0, i].axvline(mean_val, color='red', linestyle='--', alpha=0.7, label=f'Mean: {mean_val:.1f}')
    axes[0, i].axvline(median_val, color='green', linestyle='--', alpha=0.7, label=f'Median: {median_val:.1f}')
    axes[0, i].legend()
    
    # KDE Plot by Churn Status
    axes[1, i].set_title(f'{col} Density by Churn Status', fontsize=14, fontweight='bold')
    sns.kdeplot(data=df[df['Churn'] == 'No'], x=col, fill=True, color='lightgreen', 
                alpha=0.7, label='No Churn', ax=axes[1, i])
    sns.kdeplot(data=df[df['Churn'] == 'Yes'], x=col, fill=True, color='salmon', 
                alpha=0.7, label='Churn', ax=axes[1, i])
    axes[1, i].set_xlabel(col, fontsize=12)
    axes[1, i].set_ylabel('Density', fontsize=12)
    axes[1, i].legend()
    axes[1, i].grid(alpha=0.3)

plt.tight_layout()
save_path = os.path.join(figures_dir, '05_comprehensive_numeric_analysis.png')
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")

# Print insights
print("\n" + "="*60)
print("NUMERIC VARIABLES INSIGHTS")
print("="*60)
for col in numeric_cols:
    print(f"\n{col}:")
    print(f"  • Mean: {df[col].mean():.2f}")
    print(f"  • Median: {df[col].median():.2f}")
    print(f"  • Std Dev: {df[col].std():.2f}")
    print(f"  • Range: {df[col].min():.2f} - {df[col].max():.2f}")
    
    # Churn comparison
    no_churn_mean = df[df['Churn'] == 'No'][col].mean()
    churn_mean = df[df['Churn'] == 'Yes'][col].mean()
    print(f"  • No Churn avg: {no_churn_mean:.2f}")
    print(f"  • Churn avg: {churn_mean:.2f}")
    print(f"  • Difference: {churn_mean - no_churn_mean:.2f}")

plt.show()
```

```python
# 6. Box plot: MonthlyCharges by Churn
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette=['lightgreen', 'salmon'])
plt.title('Monthly Charges by Churn Status', fontsize=16, fontweight='bold')
plt.xlabel('Churn', fontsize=12)
plt.ylabel('Monthly Charges ($)', fontsize=12)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
save_path = r'D:\EDU\DAS 601 ML\Final Project\Results\figures\06_monthly_charges_by_churn.png'
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")
plt.show()
```

```python
# 7. Comprehensive Scatter Plot & Regression Analysis
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Tenure vs MonthlyCharges with regression
sns.regplot(data=df, x='tenure', y='MonthlyCharges', scatter_kws={'alpha':0.6}, 
            line_kws={'color':'red'}, ax=axes[0,0])
axes[0,0].set_title('Tenure vs Monthly Charges\n(with Regression Line)', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Tenure (months)', fontsize=12)
axes[0,0].set_ylabel('Monthly Charges ($)', fontsize=12)
axes[0,0].grid(alpha=0.3)

# 2. Tenure vs TotalCharges with regression
sns.regplot(data=df, x='tenure', y='TotalCharges', scatter_kws={'alpha':0.6}, 
            line_kws={'color':'red'}, ax=axes[0,1])
axes[0,1].set_title('Tenure vs Total Charges\n(with Regression Line)', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Tenure (months)', fontsize=12)
axes[0,1].set_ylabel('Total Charges ($)', fontsize=12)
axes[0,1].grid(alpha=0.3)

# 3. MonthlyCharges vs TotalCharges with regression
sns.regplot(data=df, x='MonthlyCharges', y='TotalCharges', scatter_kws={'alpha':0.6}, 
            line_kws={'color':'red'}, ax=axes[1,0])
axes[1,0].set_title('Monthly vs Total Charges\n(with Regression Line)', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Monthly Charges ($)', fontsize=12)
axes[1,0].set_ylabel('Total Charges ($)', fontsize=12)
axes[1,0].grid(alpha=0.3)

# 4. Scatter plot colored by Churn
scatter = axes[1,1].scatter(df['tenure'], df['MonthlyCharges'], 
                           c=df['Churn'].map({'No': 0, 'Yes': 1}), 
                           alpha=0.6, cmap='RdYlGn_r', edgecolors='black', linewidth=0.5)
axes[1,1].set_title('Tenure vs Monthly Charges\n(Colored by Churn Status)', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Tenure (months)', fontsize=12)
axes[1,1].set_ylabel('Monthly Charges ($)', fontsize=12)
cbar = plt.colorbar(scatter, ax=axes[1,1])
cbar.set_label('Churn (0=No, 1=Yes)')
axes[1,1].grid(alpha=0.3)

plt.tight_layout()
save_path = os.path.join(figures_dir, '07_comprehensive_scatter_regression.png')
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")

# Calculate and print correlation coefficients
print("\n" + "="*60)
print("CORRELATION COEFFICIENTS")
print("="*60)
correlations = [
    ('Tenure', 'MonthlyCharges', df['tenure'].corr(df['MonthlyCharges'])),
    ('Tenure', 'TotalCharges', df['tenure'].corr(df['TotalCharges'])),
    ('MonthlyCharges', 'TotalCharges', df['MonthlyCharges'].corr(df['TotalCharges']))
]

for var1, var2, corr in correlations:
    strength = "Strong" if abs(corr) > 0.7 else "Moderate" if abs(corr) > 0.3 else "Weak"
    direction = "Positive" if corr > 0 else "Negative"
    print(f"• {var1} vs {var2}: {corr:.3f} ({strength} {direction})")

plt.show()
```

```python
# 8. Bubble plot: Tenure vs MonthlyCharges sized by TotalCharges, hue=Churn
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['tenure'], df['MonthlyCharges'], 
                     s=df['TotalCharges']/20,  # Scale down for better visualization
                     c=df['Churn'].map({'No': 0, 'Yes': 1}), 
                     alpha=0.6, cmap='RdYlGn_r', edgecolors='black', linewidth=0.5)
plt.title('Tenure vs Monthly Charges\n(Bubble size = Total Charges, Color = Churn)', fontsize=16, fontweight='bold')
plt.xlabel('Tenure (months)', fontsize=12)
plt.ylabel('Monthly Charges ($)', fontsize=12)
plt.colorbar(scatter, label='Churn (0=No, 1=Yes)')
plt.grid(alpha=0.3)

# Add legend for bubble sizes
sizes = [1000, 3000, 5000]
labels = ['$1000', '$3000', '$5000']
legend_elements = [plt.scatter([], [], s=s/20, c='gray', alpha=0.6) for s in sizes]
plt.legend(legend_elements, labels, title='Total Charges', loc='upper left')

plt.tight_layout()
save_path = os.path.join(figures_dir, '08_bubble_plot_tenure_charges_churn.png')
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")
plt.show()
```

```python
# 10. Comprehensive Heatmap Analysis: Two Categorical Variables + One Numeric
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Mean Monthly Charges by Contract and Internet Service
pivot1 = df.pivot_table(values='MonthlyCharges', index='Contract', columns='InternetService', aggfunc='mean')
sns.heatmap(pivot1, annot=True, fmt='.1f', cmap='YlOrRd', ax=axes[0,0], 
            cbar_kws={'label': 'Mean Monthly Charges ($)'})
axes[0,0].set_title('Mean Monthly Charges by\nContract Type & Internet Service', fontweight='bold')

# 2. Mean Total Charges by Contract and Payment Method
pivot2 = df.pivot_table(values='TotalCharges', index='Contract', columns='PaymentMethod', aggfunc='mean')
sns.heatmap(pivot2, annot=True, fmt='.0f', cmap='Greens', ax=axes[0,1], 
            cbar_kws={'label': 'Mean Total Charges ($)'})
axes[0,1].set_title('Mean Total Charges by\nContract Type & Payment Method', fontweight='bold')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Churn Rate by Internet Service and Online Security
churn_pivot = pd.crosstab(df['InternetService'], df['OnlineSecurity'], 
                         values=df['Churn'].map({'Yes': 1, 'No': 0}), aggfunc='mean') * 100
sns.heatmap(churn_pivot, annot=True, fmt='.1f', cmap='Reds', ax=axes[1,0], 
            cbar_kws={'label': 'Churn Rate (%)'})
axes[1,0].set_title('Churn Rate (%) by\nInternet Service & Online Security', fontweight='bold')

# 4. Mean Tenure by Contract and Partner Status
pivot4 = df.pivot_table(values='tenure', index='Contract', columns='Partner', aggfunc='mean')
sns.heatmap(pivot4, annot=True, fmt='.1f', cmap='Blues', ax=axes[1,1], 
            cbar_kws={'label': 'Mean Tenure (months)'})
axes[1,1].set_title('Mean Tenure by\nContract Type & Partner Status', fontweight='bold')

plt.tight_layout()
save_path = os.path.join(figures_dir, '10_comprehensive_heatmaps.png')
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")

# Print insights
print("\n" + "="*60)
print("HEATMAP INSIGHTS")
print("n" + "="*60)
print("1. Monthly Charges Patterns:")
print(f"   • Highest: Fiber optic + Month-to-month: ${pivot1.max().max():.1f}")
print(f"   • Lowest: No internet + Two year: ${pivot1.min().min():.1f}")

print("\n2. Churn Rate Patterns:")
print(f"   • Highest risk: {churn_pivot.max().max():.1f}% (Fiber optic without security)")
print(f"   • Lowest risk: {churn_pivot.min().min():.1f}% (No internet with security)")

print("\n3. Tenure Patterns:")
print(f"   • Longest tenure: {pivot4.max().max():.1f} months (Two year + Partner)")
print(f"   • Shortest tenure: {pivot4.min().min():.1f} months (Month-to-month + No Partner)")

plt.show()
```

## 📊 Visual Analysis Following Python Graph Gallery Standards

### **Categorical Variables Visualization**

#### 1. **Bar Chart: Gender Distribution**
**Chart Type:** Simple Bar Chart  
**Purpose:** Visualize the distribution of a single categorical variable (gender)  
**Design Elements:**
- **Colors:** Sky blue for Male, Light coral for Female for clear gender distinction
- **Layout:** Vertical bars with value labels for exact counts
- **Grid:** Horizontal grid lines for easy value reading

**📈 Key Insights:**
- **Balanced Distribution:** Nearly equal representation (Male: 2,823 vs Female: 2,802)
- **Business Implication:** Gender-neutral marketing strategies should be effective
- **Sample Adequacy:** No gender bias in data collection

---

#### 2. **Grouped Bar Chart: Contract Type vs Churn**
**Chart Type:** Grouped Bar Chart  
**Purpose:** Compare churn rates across different contract types  
**Design Elements:**
- **Colors:** Light green for retention (positive), Salmon for churn (negative)
- **Layout:** Side-by-side bars for direct comparison
- **Legend:** Clear labeling of churn status

**📈 Key Insights:**
- **Month-to-month Risk:** Highest absolute churn numbers (1,346 churned vs 1,749 retained)
- **Contract Loyalty:** Two-year contracts show strongest retention (only 62 churned vs 1,306 retained)
- **Business Priority:** Focus retention efforts on month-to-month customers

---

#### 3. **Stacked Bar Charts: Internet Service vs Churn**
**Chart Type:** Stacked Bar Chart & 100% Stacked Bar Chart  
**Purpose:** Show both absolute numbers and proportional churn rates by service type  
**Design Elements:**
- **Dual View:** Absolute counts (left) and percentages (right)
- **Color Consistency:** Same green/salmon scheme for easy interpretation
- **Proportional Insight:** 100% stacked reveals true churn rates

**📈 Key Insights:**
- **Fiber Optic Challenge:** ~40% churn rate despite larger customer base
- **DSL Stability:** ~20% churn rate shows better customer satisfaction

```python
# 14. Comprehensive Correlation Analysis
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Encode categorical variables for correlation analysis
df_encoded = df.copy()

# Encode binary categorical variables
binary_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in binary_cols:
    df_encoded[col] = df_encoded[col].map({'Male': 1, 'Female': 0, 'Yes': 1, 'No': 0}).fillna(df_encoded[col])

# Encode multi-category variables using label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

categorical_cols = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
                   'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
                   'Contract', 'PaymentMethod']

for col in categorical_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

# Select key columns for correlation
key_cols = ['SeniorCitizen', 'Partner', 'Dependents', 'tenure', 'InternetService', 
           'OnlineSecurity', 'TechSupport', 'Contract', 'PaperlessBilling', 
           'MonthlyCharges', 'TotalCharges', 'Churn']

corr_matrix = df_encoded[key_cols].corr()

# 1. Main correlation heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, square=True, 
            fmt='.3f', ax=axes[0,0], cbar_kws={'shrink': 0.8})
axes[0,0].set_title('Feature Correlation Matrix', fontweight='bold', fontsize=14)
axes[0,0].tick_params(axis='x', rotation=45)

# 2. Correlation with Churn
churn_corr = corr_matrix['Churn'].drop('Churn').sort_values(key=abs, ascending=False)
colors = ['darkred' if x < -0.2 else 'red' if x < 0 else 'darkgreen' if x > 0.2 else 'green' for x in churn_corr.values]
bars = axes[0,1].barh(range(len(churn_corr)), churn_corr.values, color=colors, alpha=0.7)
axes[0,1].set_yticks(range(len(churn_corr)))
axes[0,1].set_yticklabels(churn_corr.index)
axes[0,1].set_xlabel('Correlation with Churn')
axes[0,1].set_title('Feature-Churn Correlations', fontweight='bold', fontsize=14)
axes[0,1].grid(axis='x', alpha=0.3)
axes[0,1].axvline(0, color='black', linestyle='-', alpha=0.5)

# Add value labels on bars
for i, (bar, value) in enumerate(zip(bars, churn_corr.values)):
    axes[0,1].text(value + (0.01 if value >= 0 else -0.01), i, f'{value:.3f}', 
                   va='center', ha='left' if value >= 0 else 'right', fontweight='bold')

# 3. Numeric variables correlation
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
numeric_corr = df[numeric_cols].corr()
sns.heatmap(numeric_corr, annot=True, cmap='coolwarm', center=0, square=True, 
            fmt='.3f', ax=axes[1,0], cbar_kws={'shrink': 0.8})
axes[1,0].set_title('Numeric Variables Correlation', fontweight='bold', fontsize=14)

# 4. Service features correlation with churn
service_cols = ['OnlineSecurity', 'TechSupport', 'OnlineBackup', 'DeviceProtection', 
               'StreamingTV', 'StreamingMovies']
service_churn_corr = df_encoded[service_cols + ['Churn']].corr()['Churn'].drop('Churn')
service_churn_corr = service_churn_corr.sort_values(key=abs, ascending=False)

colors = ['darkred' if x < -0.15 else 'red' if x < 0 else 'darkgreen' if x > 0.15 else 'green' for x in service_churn_corr.values]
bars = axes[1,1].barh(range(len(service_churn_corr)), service_churn_corr.values, color=colors, alpha=0.7)
axes[1,1].set_yticks(range(len(service_churn_corr)))
axes[1,1].set_yticklabels(service_churn_corr.index)
axes[1,1].set_xlabel('Correlation with Churn')
axes[1,1].set_title('Service Features vs Churn', fontweight='bold', fontsize=14)
axes[1,1].grid(axis='x', alpha=0.3)
axes[1,1].axvline(0, color='black', linestyle='-', alpha=0.5)

plt.tight_layout()
save_path = os.path.join(figures_dir, '13_comprehensive_correlation_analysis.png')
plt.savefig(save_path, dpi=150, bbox_inches='tight')
print(f"Saved: {save_path}")

# Print detailed insights
print("\n" + "="*80)
print("COMPREHENSIVE CORRELATION ANALYSIS INSIGHTS")
print("="*80)

print("\n🔴 TOP CHURN RISK FACTORS (Positive Correlation):")
positive_churn = churn_corr[churn_corr > 0].sort_values(ascending=False)
for i, (feature, corr) in enumerate(positive_churn.head(3).items(), 1):
    impact = "High" if corr > 0.15 else "Moderate" if corr > 0.1 else "Low"
    print(f"  {i}. {feature}: +{corr:.3f} ({impact} Impact)")

print("\n🟢 TOP RETENTION FACTORS (Negative Correlation):")
negative_churn = churn_corr[churn_corr < 0].sort_values(ascending=True)
for i, (feature, corr) in enumerate(negative_churn.head(3).items(), 1):
    impact = "High" if abs(corr) > 0.15 else "Moderate" if abs(corr) > 0.1 else "Low"
    print(f"  {i}. {feature}: {corr:.3f} ({impact} Impact)")

print("\n📊 STRONGEST FEATURE RELATIONSHIPS:")
strongest_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if corr_matrix.columns[i] != 'Churn' and corr_matrix.columns[j] != 'Churn':
            corr_val = corr_matrix.iloc[i, j]
            if abs(corr_val) > 0.3:
                strongest_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_val))

strongest_pairs.sort(key=lambda x: abs(x[2]), reverse=True)
for i, (feat1, feat2, corr) in enumerate(strongest_pairs[:3], 1):
    relationship = "Strong Positive" if corr > 0.5 else "Moderate Positive" if corr > 0.3 else "Strong Negative" if corr < -0.5 else "Moderate Negative"
    print(f"  {i}. {feat1} ↔ {feat2}: {corr:.3f} ({relationship})")

print("\n💡 SERVICE STRATEGY INSIGHTS:")
print("• Security services (OnlineSecurity, TechSupport) show strongest retention correlation")
print("• Streaming services have minimal impact on churn")
print("• Device protection and backup services provide moderate retention value")

plt.show()
```
